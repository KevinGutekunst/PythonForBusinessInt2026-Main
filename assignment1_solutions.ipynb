### Task 1: A list of dictionaries (1p)

You are given a list of raw transaction records, where each record is a dictionary:

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

Write a script that performs the following steps:

a) Filter out invalid transactions where the price is zero or the quantity is negative.

b) For each remaining transaction, compute a new field `total` = `price * quantity`.

c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

d) Print the region with the highest total revenue and its value using an f-string.

````python
# --- Task 1: A list of dictionaries ---

transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# a) Filter out invalid transactions (price == 0 or quantity < 0)
valid = []
for t in transactions:
    if t["price"] > 0 and t["quantity"] > 0:
        valid.append(t)

# b) Compute total for each valid transaction
for t in valid:
    t["total"] = t["price"] * t["quantity"]

# c) Group by region and compute total revenue per region
revenue_by_region = {}
for t in valid:
    region = t["region"]
    if region not in revenue_by_region:
        revenue_by_region[region] = 0
    revenue_by_region[region] += t["total"]

print("Revenue by region:", revenue_by_region)

# d) Print the region with the highest total revenue
best_region = max(revenue_by_region, key=revenue_by_region.get)

print(f"Highest revenue region: {best_region} with {revenue_by_region[best_region]}€")
````

### Task 2: Creating a function (1p)

You are given the following list of daily sales figures (in euros) for a two-week period:

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

Write a function called `summarise_sales` that takes a list of daily sales values and returns a dictionary containing:

`"total"`: the sum of all daily sales

`"average"`: the mean daily sales value, rounded to two decimals

`"best_day"`: the index (starting from 1) of the day with the highest sales

`"worst_day"`: the index (starting from 1) of the day with the lowest sales

`"above_average_count"`: the number of days where sales exceeded the average

Call the function with the list above and print the returned dictionary.

````python
# --- Task 2: Creating a function ---

sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(daily_sales):
    """Return a summary dictionary for a list of daily sales values."""
    total = sum(daily_sales)
    average = round(total / len(daily_sales), 2)
    best_day = daily_sales.index(max(daily_sales)) + 1   # 1-indexed
    worst_day = daily_sales.index(min(daily_sales)) + 1   # 1-indexed
    above_average_count = sum(1 for s in daily_sales if s > average)

    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count,
    }

result = summarise_sales(sales)
print(result)
````

### Task 3: Loops (1p)

You are given the following list of student records:

```python
students = [

    {"name": "Erik", "grade": 58, "attendance": 0.75},

    {"name": "Tom", "grade": 67, "attendance": 0.70},

    {"name": "Olli", "grade": 50, "attendance": 0.65},

```

Using `for` loops and `if` statements, perform the following:

a) Create a new list called `passing` that contains only the names of students with a grade of 50 or above. Print the list.

b) Loop through the students and print a status message for each. If the student's grade is 80 or above, print `"[name]: Excellent"`; if between 50 and 79, print `"[name]: Pass"`; otherwise print `"[name]: Fail"`.

c) Count how many students have an attendance rate below 0.80. Print the count.

d) Find the student with the highest grade using a loop (do not use `max()`). Print the student's name and grade.

e) Now find the student with the highest grade using `max()`. Print the student's name and grade.

````python
# --- Task 3: Loops ---

students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# a) List of passing students (grade >= 50)
passing = [s["name"] for s in students if s["grade"] >= 50]
print("Passing:", passing)

# b) Status message for each student
for s in students:
    if s["grade"] >= 80:
        print(f"{s['name']}: Excellent")
    elif s["grade"] >= 50:
        print(f"{s['name']}: Pass")
    else:
        print(f"{s['name']}: Fail")

# c) Count students with attendance below 0.80
low_attendance = sum(1 for s in students if s["attendance"] < 0.80)
print(f"\nStudents with attendance below 0.80: {low_attendance}")

# d) Find the student with the highest grade using a loop (no max())
best_student = students[0]
for s in students[1:]:
    if s["grade"] > best_student["grade"]:
        best_student = s
print(f"\nHighest grade (loop): {best_student['name']} with grade {best_student['grade']}")

# e) Find the student with the highest grade using max()
best_student_max = max(students, key=lambda s: s["grade"])

print(f"Highest grade (max): {best_student_max['name']} with grade {best_student_max['grade']}")
````

### Task 4: Class and methods (1p)

Design a class called `Portfolio` that represents an investment portfolio. The constructor is provided for you:

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
```

Each key in `holdings` is a ticker symbol (string) and each value is a dictionary with keys `"shares"` and `"price_per_share"`. Add the following methods to the class:

a) A method `add_asset(ticker, shares, price_per_share)` that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).

b) A method `total_value()` that returns the total market value of the portfolio.

c) A method `most_valuable()` that returns the ticker of the asset with the highest total value (shares × price).

d) A `__str__` method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.

Create an instance, add at least three assets, and demonstrate all methods with printed output.

````python
# --- Task 4: Class and methods ---

class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    # a) Add or update an asset
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            self.holdings[ticker] = {"shares": shares, "price_per_share": price_per_share}

    # b) Total market value of the portfolio
    def total_value(self):
        return sum(h["shares"] * h["price_per_share"] for h in self.holdings.values())

    # c) Ticker of the most valuable asset
    def most_valuable(self):
        return max(
            self.holdings,
            key=lambda t: self.holdings[t]["shares"] * self.holdings[t]["price_per_share"],
        )

    # d) Readable summary
    def __str__(self):
        return (
            f"Portfolio of {self.owner}: "
            f"{len(self.holdings)} asset(s), "
            f"total value = {self.total_value():.2f}€"
        )

# Demonstration
p = Portfolio("Alice")
p.add_asset("AAPL", 10, 175.50)
p.add_asset("MSFT", 5, 410.00)
p.add_asset("NVDA", 8, 890.25)

print(p)
print(f"Total value: {p.total_value():.2f}€")
print(f"Most valuable asset: {p.most_valuable()}")
````

### Task 5: Importing a module (1p)

You are given the following list of order dates (as strings) and product names:

```python
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
```

Using the `datetime` module, parse each date string into a `datetime.date` object. Compute and print the number of days between the earliest and latest order.

````python
# --- Task 5: Importing a module ---

from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

# Parse date strings into date objects
dates = []
for order in orders:
    date_string = order[0]
    date_object = datetime.strptime(date_string, "%Y-%m-%d").date()
    dates.append(date_object)
    
# Compute the span between earliest and latest order
earliest = min(dates)
latest = max(dates)
span = (latest - earliest).days

print(f"Earliest order: {earliest}")
print(f"Latest order:   {latest}")
print(f"Days between earliest and latest order: {span}")
````

### Task 6: Array Operations and Broadcasting (1p)

A company tracks quarterly revenue (in thousands of euros) across four quarters over three years:

```python
# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])
```

a) Compute and print the total annual revenue for each year (i.e., row sums). Use the `axis` parameter.

b) Compute the mean quarterly revenue across all years (i.e., column means).

c) Using broadcasting, apply a growth adjustment to the 2024 row by multiplying it element-wise with the array `[1.05, 1.03, 1.08, 1.02]`. Print the adjusted values.

d) Use boolean indexing to find and print all quarterly revenue values that exceed 250 (thousands of euros).

````python
# --- Task 6: Array Operations and Broadcasting ---

import numpy as np

# Rows: years (2022, 2023, 2024), Columns: quarters (Q1–Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310],
])

# a) Total annual revenue for each year (row sums)
annual_totals = revenue.sum(axis=1)
print("Annual revenue:", annual_totals)

# b) Mean quarterly revenue across all years (column means)
quarterly_means = revenue.mean(axis=0)
print("Mean quarterly revenue:", quarterly_means)

# c) Apply growth adjustment to the 2024 row via broadcasting
adjustment = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * adjustment
print("Adjusted 2024 revenue:", adjusted_2024)

# d) Boolean indexing: quarterly values exceeding 250 (thousands €)
above_250 = revenue[revenue > 250]
print("Quarters exceeding 250k€:", above_250)
````

Create the following DataFrame:

data = {

````python
# --- Task 7: Pandas ---

import pandas as pd

data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9],
}
df = pd.DataFrame(data)

# a) New column: salary per year of experience (rounded to integer)
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round(0).astype(int)
print("With salary_per_year_exp:")
print(df, "\n")
````

````python
# c) Finance employees with performance score above 4.0
finance_high = df.loc[(df["department"] == "Finance") & (df["performance_score"] > 4.0)]
print("Finance department, score > 4.0:")
print(finance_high, "\n")
````

````python
# d) Rename years_exp to experience_years
df = df.rename(columns={"years_exp": "experience_years"})
print("Renamed columns:", list(df.columns), "\n")
````

````python
# e) Salary and performance score for the first four employees (iloc)
subset = df.iloc[:4, [2, 4]]  # salary is col 2, performance_score is col 4
print("First four employees – salary & performance score:")
print(subset)
````

### Task 8: More Pandas (1p)

A company records daily website traffic over four weeks. Generate the data by running the following script first:

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

Using this generated data, perform the following:

a) Create a DataFrame with columns: `date`, `visitors`, `conversion_rate`, and `avg_order_value`. Set `date` as the index.

b) Compute two new columns: `conversions` (= visitors × conversion_rate, rounded to the nearest integer) and `daily_revenue` (= conversions × avg_order_value).

````python
# --- Task 8: More Pandas ---

import numpy as np
import pandas as pd

# Generate data with the given seed
rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

# a) Create DataFrame with date as index
traffic = pd.DataFrame({
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value,
}, index=days)
traffic.index.name = "date"

print(traffic.head(), "\n")
````

````python
# b) Compute conversions and daily revenue
traffic["conversions"] = (traffic["visitors"] * traffic["conversion_rate"]).round(0).astype(int)
traffic["daily_revenue"] = traffic["conversions"] * traffic["avg_order_value"]

print(traffic.head())
````
