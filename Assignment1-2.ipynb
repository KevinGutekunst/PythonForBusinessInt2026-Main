```python
#Task 1

transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
#a)
valid_transactions = []

for t in transactions:
    if t["price"] > 0 and t["quantity"] >=0:
        valid_transactions.append(t)

#b)
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]

#c)
region_totals = {}

for t in valid_transactions:
    region = t["region"]

    if region not in region_totals:
        region_totals[region] = 0

    region_totals[region] += t["total"]

#d)
max_region = ""
max_value = 0

for region in region_totals:
    if region_totals[region] > max_value:
        max_value = region_totals[region]
        max_region = region

print(f"Region with highest revenue: {max_region} ({max_value})")
```

```python
#Task 2

def summarise_sales(sales):

    # total
    total = sum(sales)

    # average
    average = round(total / len(sales), 2)

    # best day and worst day
    best_day = sales.index(max(sales)) + 1
    worst_day = sales.index(min(sales)) + 1

    # above average count
    count = 0
    for s in sales:
        if s > average:
            count +=1

    # return dictionary
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": count
    }

sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

result = summarise_sales(sales)

print(result)
```

```python
#Task 3

students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

#a)

passing = []

for student in students:
    if student["grade"] >=50:
        passing.append(student["name"])

print(passing)

#b)

for student in students:
    name = student["name"]
    grade = student["grade"]
    
    if grade >= 80:
        print(f"{name}: Excellent")
    elif grade >= 50:
        print(f"{name}: Pass")
    else:
        print(f"{name}: Fail")

#c)

count = 0

for student in students:
    if student["attendance"] < 0.80:
        count += 1

print(count)

#d)
top_student = None
top_grade = 0

for student in students:
    if student["grade"] > top_grade:
        top_grade = student["grade"]
        top_student = student["name"]

print(f"{top_student}: {top_grade}")

#e)

best = max(students, key=lambda x: x["grade"])

print(f"{best['name']}: {best['grade']}")
```

```python
#Task 4

class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
    
    # a) 
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            # update shares and price
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            # add new asset
            self.holdings[ticker] = {
                "shares": shares,
                "price_per_share": price_per_share
            }
    
    # b) 
    def total_value(self):
        total = 0
        for ticker in self.holdings:
            asset = self.holdings[ticker]
            total += asset["shares"] * asset["price_per_share"]
        return total
    
    # c) 
    def most_valuable(self):
        max_value = 0
        best_ticker = ""
        
        for ticker in self.holdings:
            asset = self.holdings[ticker]
            value = asset["shares"] * asset["price_per_share"]
            
            if value > max_value:
                max_value = value
                best_ticker = ticker
        
        return best_ticker
    
    # d) 
    def __str__(self):
        num_assets = len(self.holdings)
        total = self.total_value()
        
        return f"Owner: {self.owner}, Assets: {num_assets}, Total Value: {total}"

portfolio = Portfolio("Kevin")

portfolio.add_asset("AAPL", 10, 150)
portfolio.add_asset("GOOG", 5, 1000)
portfolio.add_asset("TSLA", 8, 200)

print(portfolio.total_value())
print(portfolio.most_valuable())
print(portfolio)
```

```python
#Task 5
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
from datetime import datetime

dates = []

for order in orders:
    date_string = order[0]
    date_object = datetime.strptime(date_string, "%Y-%m-%d").date()
    dates.append(date_object)

earliest = min(dates)
latest = max(dates)

difference = (latest - earliest).days

print(difference)
```

```python
#Task 8

import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

#a)
df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

df = df.set_index("date")

print(df.head())

#b)
df["conversions"] = (df["visitors"] * df["conversion_rate"]).round()

df["daily_revenue"] = df["conversions"] * df["avg_order_value"]

print(df.head())
```
